```python
# WRITE QUERIES for questions

sql01 = '''select distinct trim(m.title) as Movie_name,trim(p.name) as Director_name, trim(m.year) as Year from
Genre g join m_genre mg on trim(g.gid)= trim(mg.gid)
 join Movie m on trim(mg.mid) = trim(m.mid) 
  join M_Director md on trim(md.mid) = trim(m.mid)
   join Person p on trim(p.pid) = trim(md.pid)
 where trim(g.name)='Comedy' and (((m.year%4)=0 and (m.year%100) <> 0) or (m.year%400)=0);'''
%time t13 = pd.read_sql_query(sql01,conn)  #gives calculaion time
pd.read_sql_query(sql01,conn)
```

```python
# GENERIC SQLITE CONNECTION
sql02 = '''Select p.name from person p join M_cast c on trim(p.pid)=trim(c.pid)
inner join Movie m on c.mid = m.mid where m.title='Anand';'''
%time t13 = pd.read_sql_query(sql02,conn)  #gives calculaion time
pd.read_sql_query(sql02,conn)
```

```python
# GENERIC SQLITE CONNECTION
sql03 = '''Select
    p.name
from
    person p
    inner join M_cast c on trim(p.pid) = trim(c.pid)
    inner join Movie m on c.mid = m.mid
where
    m.year not between 1970
    and 1990 limit 1000
'''
%time t13 = pd.read_sql_query(sql03,conn)  #gives calculaion time
pd.read_sql_query(sql03,conn) 
```

```python
# GENERIC SQLITE CONNECTION
sql04 = '''select
    trim(p.name) as Director_name,
    count(m.mid) as No_of_Movies
from
    Movie m
    inner join M_Director md on trim(md.mid) = trim(m.mid)
    inner join Person p on trim(p.pid) = trim(md.pid)
group by
    trim(p.name)
having
    count(m.mid) > 10
order by
    count(m.mid) DESC'''
%time t13 = pd.read_sql_query(sql04,conn)  #gives calculaion time
pd.read_sql_query(sql04,conn) 
```

```python
# GENERIC SQLITE CONNECTION
sql05a = '''select
    trim(m.year) as Year,
    count(m.mid) as No_of_Movies_by_Female
from
    Movie m
    inner join M_Director md on trim(md.mid) = trim(m.mid)
    inner join Person p on trim(p.pid) = trim(md.pid)
where
    p.gender = 'Female'
group by
    trim(m.year)
order by
    m.year DESC limit 100'''
%time t13 = pd.read_sql_query(sql05a,conn)  #gives calculaion time
pd.read_sql_query(sql05a,conn)
```

```python
# GENERIC SQLITE CONNECTION
sql05b = '''Select 
Year,
((No_of_Movies_by_Female/(select count(m.mid) from Movie m where m.year = X.year ))*100) as Movie_by_Females,
(select count(m.mid) from Movie m where m.year = X.year) as Total_no_of_Movies
from (
select
    trim(m.year) as Year,
    count(m.mid) as No_of_Movies_by_Female
from
    Movie m
    inner join M_Director md on trim(md.mid) = trim(m.mid)
    inner join Person p on trim(p.pid) = trim(md.pid)
where
    trim(p.gender) = 'Female'
group by
    trim(m.year)
order by
    m.year DESC
	limit 10) X
    limit 10 '''
%time t13 = pd.read_sql_query(sql05b,conn)  #gives calculaion time
pd.read_sql_query(sql05b,conn)
```

```python
# GENERIC SQLITE CONNECTION
sql06 = '''Select
    m.title as Title,count(c.pid) as Cast_Size
from M_cast c inner join Movie m on c.mid = m.mid group by c.mid order by count(c.pid) desc limit 1'''
%time t13 = pd.read_sql_query(sql06,conn)  #gives calculaion time
pd.read_sql_query(sql06,conn)
```

```python
# GENERIC SQLITE CONNECTION
sql07 = '''select y.year as decade_start, y.year + 9 as decade_end,
       count(mid) as num_movies
from (select distinct year from movie) y join
     movie m
     on m.year >= y.year and m.year < y.year + 10
group by y.year
order by count(mid) desc
limit 1;'''
%time t13 = pd.read_sql_query(sql07,conn)  #gives calculaion time
pd.read_sql_query(sql07,conn)
```

```python
# GENERIC SQLITE CONNECTION
sql08 = '''Select PID, Name from Person where PID not in
(select distinct(PID) from M_Cast as C1 natural join Movie as M1 where exists(select MID from M_Cast as C2 natural join Movie as M2 
where C1.PID = C2.PID and (M2.year - 3) > M1.year 
and not exists(select MID from M_Cast as C3 natural join Movie as M3 where C1.PID = C3.PID and M1.year < M3.year and M3.year < M2.year)));'''
%time t13 = pd.read_sql_query(sql08,conn)  #gives calculaion time
pd.read_sql_query(sql08,conn)
```
